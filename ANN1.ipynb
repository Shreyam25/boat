ACTIVATION FUNCTIONS IN DEEP LEARNING

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)

x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
# plt.axis('tight')
plt.title('Activation Function : BinaryStep')
plt.show()
```

```python
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x

x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function : Linear')
plt.show()
```

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function : Sigmoid')
plt.show()
```

```python

def relu(x):
	return max(0.0, x)

input = [x for x in range(-5, 10)]
# apply relu on each input
output = [relu(x) for x in input]
# plot our result
plt.plot(input,output)
plt.show()
```

```python
def Leaky_RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0.01*i)
        else:
            x1.append(i)

    return x1

x = np.linspace(-10, 10)
plt.plot(x, Leaky_RELU(x))
plt.axis('tight')
plt.title('Activation Function : LEAKY RELU')
plt.show()
```
